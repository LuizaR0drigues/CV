### Manipulando imagens com OpenCV

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageDraw
import numpy as np
import cv2
```

```python
#Imagem Original
image = cv2.imread("./img/cat.png")
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Cortando Imagens

a variável <code>upper</code> é a primeira linha que gostaríamos de incluir na imagem, a variável <code>lower</code> é a última linha que gostaríamos de incluir. Então usamos o slicing para obter a nova imagem.

```python
upper = 150
lower = 400

image = cv2.imread("./img/cat.png")
crop_top = image[upper:lower,:,:]
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(crop_top, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
left = 150
right = 400
crop_horizontal =crop_top[:,left:right,:]
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(crop_horizontal, cv2.COLOR_BGR2RGB))
plt.show()
```

### Modificando usando pixels e canais

```python
#alterando img
quadrado = np.copy(image)

#upper Margem superior
#lower Margem inferior
#left Margem esquerda
#right Margem direita

# Altere a área ao redor do quadrado para preto.
quadrado[:upper, :, :] = 0  # Parte superior
quadrado[lower:, :, :] = 0  # Parte inferior
quadrado[:, :left, :] = 0  # Parte esquerda
quadrado[:, right:, :] = 0  # Parte direita

# Exibir as imagens.
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(quadrado, cv2.COLOR_BGR2RGB))
plt.title("Altered Image")
plt.show()
```

Também podemos criar formas e <code>OpenCV</code>, podemos usar o método <code>rectangle</code>. O parâmetro <code>pt1</code> é a coordenada superior esquerda do retângulo: <code>(left,top)</code> ou $(x_0,y_0)$, <code>pt2</code> é a coordenada inferior direita<code>(right,lower)</code> ou $(x_1,y_1)$. O parâmetro <code>color</code> é uma tupla que representa a intensidade de cada canal <code>(blue, green, red)</code>. Finalmente, temos a espessura da linha.

```python
ponto_inicial, ponto_final = (left, upper), (right, lower)
image_draw = np.copy(image)
#funcao rectangle 
#esta funcao usao o padrao bgr
cv2.rectangle(image_draw, pt1 = ponto_inicial, pt2=ponto_final, color =(0,0,255), thickness=3)
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(image_draw, cv2.COLOR_BGR2RGB))
plt.show()
```

adicionando textos na imagem


 <li><code>img</code>: Image array </li>
<li><code>text</code>: Text string to be overlayed</li>
<li><code>org</code>: Bottom-left corner of the text string in the image</li>
<li><code>fontFace</code>: tye type of font </li>
<li><code>fontScale</code>: Font scale</li>
<li><code>color</code>: Text color</li>
<li><code>thickness</code>: Thickness of the lines used to draw a text</li>
<li><code>lineType:</code> Line type</li>


```python
image_font = cv2.putText(img=image_draw,text='Cat',org=(150,150),color=(0,0,0),fontFace=4,fontScale=2,thickness=4)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image_draw,cv2.COLOR_BGR2RGB))
plt.show()

#salvando a img
cv2.imwrite("./img/cat.marc.jpg", image_font)
```
